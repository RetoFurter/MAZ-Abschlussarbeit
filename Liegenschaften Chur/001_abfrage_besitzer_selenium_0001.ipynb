# alle csv-Dateien in ein Pandaframe einfügen, abspeichern


```python
import pandas as pd
import glob
from tqdm import tqdm


# Liste aller csv
csv_files = glob.glob('grundstuecke/' + '*.csv')
len(csv_files)
```

```python
# CSV öffnen, in Liste alle_dfs, in csv abspeichern

alle_dfs = []
for csv_file in tqdm(csv_files):
    dfs = pd.read_csv(csv_file, sep=';')
    
    alle_dfs.append(dfs)

    
```

```python
alle_daten = pd.concat(alle_dfs)

# csv abspeichern
file = pd.DataFrame(alle_daten)            
file.to_csv('Grundstuecke_Chur.csv', sep=";", index=False)
```
